## For loops

**While loops** vs **For loops**

- while loop are used when the iterations depend on some condition.
- for loop are used when the iterations depend on the elements of a collection.

```python
# While loop
target = 7
match_found = False

while (not match_found):
    num = int(input("Enter a number: "))
    if (num == target):
        print(f"{num} is the target.")
        match_found = True
    else:
        print(f"{num} is not the target.")

print("Loop stopped!")
```

```python
# For loop on string
myString = "Loop"

# for(int i=0; i<len(myString); i++){
#     print(myString[i])
# }

for ch in myString[1:3]:
    print(f"I am currently looking at {ch}")

print("Loop stopped!")
```

```python
# For loop on list
myList = [8, 1, 4, 2, 0]

for num in myList:
    print(f"I am currently looking at {num}")

print("Loop stopped!")
```

```python
# For loop on dictionary
myDict = {'Eng': 5, 'Mat': 2, 'Sci': 1, 'SSt': 4}

for key in myDict:
    print(f"I am currently looking at {myDict[key]}")

print("Loop stopped!")
```

```python
# Running a loop for specified number of times
N = 4

for _ in range(N):
    print("Hello")

print("Loop stopped!")
```

## Nested Structures

- We can store lists or dictionaries inside other lists or dictionaries
- Structures liked these are called **nested structures**

```python
# list of lists

nested = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
print(f"type of structure = {type(nested)}")
print(f"type of elements  = {type(nested[0])}")
print()

print(nested)
print(nested[1])
print(nested[1][2])
```

```python
# list of dicts

nested = [
    {'name':'Ronaldo', 'country':'Por', 'age':40},
    {'name':'Messi',   'country':'Arg', 'age':37},
    {'name':'Neymar',  'country':'Brz', 'age':30},
    {'name':'Ramos',   'country':'Spn', 'age':39},
]
print(f"type of structure = {type(nested)}")
print(f"type of elements  = {type(nested[0])}")
print()

print(nested)
print(nested[1])
print(nested[1]['name'])
```

```python
# dict of lists

nested = {
    'names' : ['Ronaldo', 'Messi', 'Neymar', 'Ramos'],
    'countries' : ['Por', 'Arg', 'Brz', 'Spn'],
    'ages' : [40, 37, 30, 39]
}
print(f"type of structure = {type(nested)}")
print(f"type of elements  = {type(nested['names'])}")
print()

print(nested)
print(nested['names'])
print(nested['names'][2])
```

```python
# dict of dicts

nested = {
    'left_forward' : {
        'name':'Ronaldo', 'country':'Portugal', 'age':40
    },
    'right_forward' : {
        'name':'Messi', 'country':'Argentina', 'age':37
    },
    'mid_fielder' : {
        'name':'Neymar', 'country':'Brazil', 'age':30
    },
    'center_back' : {
        'name':'Ramos', 'country':'Spain', 'age':39
    }
}
print(f"type of structure = {type(nested)}")
print(f"type of elements  = {type(nested['left_forward'])}")
print()

print(nested)
print(nested['left_forward'])
print(nested['left_forward']['name'])
```

## Function

A function is control block in programming language that performs a specific task.

```python
def greet():
    print("Hello")

greet()
greet()
greet()
```

- A function can have zero, one or more parameters (inputs)
- A function can have zero, one or more return values (outputs)

```python
# NO parameter, NO return value

def greet():
    print("Hello")

x = greet()
print(x)
```

```python
# ONE parameter, NO return value

def greet(name):
    print(f"Hello {name}")

x = greet("Rancho")
print(x)
```

```python
# NO parameter, ONE return value

def greet():
    print(f"Hello")
    return 7

x = greet()
print(x)
```

```python
# ONE parameter, ONE return value
 
def greet(name):
    print(f"Hello {name}")
    return 7

x = greet("Rancho")
print(x)
```

- Function to find the greater of the two numbers

```python
def max(a, b):
    if (a > b):
        return a
    else:
        return b
    
x = max(10, 20)
print(f"ans = {x}")

x = max(20, 10)
print(f"ans = {x}")
```
